```python
import cv2

def open_image(image_path : str) -> np.ndarray:
    img = cv2.imread(image_path)
    return img, img.shape[0], img.shape[1]


def divide_image(img : np.ndarray, PLOT_CANALS : bool = False) -> np.ndarray:
    b, g, r = cv2.split(img)
    if PLOT_CANALS:
        plt.figure(figsize=(8, 5))
        plt.imshow(b, cmap='gray')
        plt.title('Blue channel')
        plt.show()
       
        plt.figure(figsize=(8, 5))
        plt.imshow(g, cmap='gray')
        plt.title('Green channel')
        plt.show()
        
        plt.figure(figsize=(8, 5))
        plt.imshow(r, cmap='gray')
        plt.title('Red channel')
        plt.show()
        
    return b, g, r
    
def image_hsv(img : np.ndarray, PLOT_CONVERSION :bool = False) -> np.ndarray:
    hsv_img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
    if PLOT_CONVERSION :
        # Display the original and HSV images side by side
        plt.figure(figsize=(10, 5))
        plt.subplot(1, 2, 1)
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.title('Original Image')
        plt.subplot(1, 2, 2)
        plt.imshow(cv2.cvtColor(hsv_img, cv2.COLOR_HSV2RGB))
        plt.title('HSV Image')

    plt.show()
    return hsv_img
```

```python
def count_pixels(img : np.ndarray) -> int:
    height = img.shape[0]
    width = img.shape[1]
    channel = img.shape[2]
    return height * width * channel
```

```python
my_image = open_image('figs\\car.png')
print(f'The image contains : {count_pixels(my_image[0])} pixels')
```

```python
def fact_pix(image : np.ndarray, factor1, factor2):
    # Cration d'une copie de l'image originale
    I_fact = image.copy()

    # Modification des intensits en fonction des facteurs
    I_fact[:, :, 0] = image[:, :, 0] * factor1
    I_fact[:, :, 1] = image[:, :, 1] * factor2

    return I_fact
```

```python
bird_image , _ , _= open_image('figs\\red_bird.png')
I_fact = fact_pix(bird_image, 1.5, 0.5)
plt.figure(figsize=(8, 5))
plt.imshow(I_fact)
```

```python
def func_a(image : np.ndarray):
    log_image = np.log1p(image).astype(np.float32)
    expo_image = np.exp(image).astype(np.float32)
    square_image = np.square(image).astype(np.float32)
    sqrt_image = np.sqrt(image).astype(np.float32)
    return log_image, expo_image, square_image, sqrt_image
```

```python
log_image, expo_image, square_image, sqrt_image = func_a(bird_image)
plt.figure(figsize=(8, 5))
plt.imshow(log_image)
plt.title('Logarithm of the image')
plt.imshow(expo_image)
plt.title('Exponential of the image')
plt.imshow(square_image)
plt.title('Square of the image')
plt.imshow(sqrt_image)
plt.title('Square root of the image')
```

```python
def func_m(image : np.ndarray):
    mean_image = np.mean(image)
    std_image = np.std(image)
    return mean_image, std_image
```

```python
def normalize(image : np.ndarray, new_min, new_max):
    norm_image = cv2.normalize(image, None, alpha=new_min, beta=new_max, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
    return norm_image
    
```

```python
norm_image = normalize(bird_image, 0, 29)
plt.figure(figsize=(8, 5))
plt.imshow(norm_image)
```

```python
def inverse_image(image : np.ndarray):
    max_intensity = np.max(image)
    inv_image = max_intensity - image
    return inv_image 
```

```python
inv_image = inverse_image(bird_image)
plt.figure(figsize=(8, 5))
plt.imshow(inv_image)
```

```python
def calc_hist(image: np.ndarray):
    # convert image to grayscale if it's in RGB
    if len(image.shape) > 2:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        # we can use np.ravel() or np.flatten() to convert the 2D array to 1D
    H, b = np.histogram(image.flatten(), 256, [0, 256])
    return H, b
```

```python
histogram, bins = calc_hist(bird_image)
plt.figure(figsize=(10, 8))
plt.plot(histogram, color='blue')
plt.title("Histogramme de l'image")
plt.xlabel("Intensit des pixels")
plt.ylabel("Frquence")
plt.show()
```

```python
def threshold(image : np.ndarray, threshold: int, PLOT_THRESH_IMG : bool = False):
    # convert image to grayscale if it's in RGB
    if len(image.shape) > 2:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    retval, thresh_img = cv2.threshold(image, threshold, 255, cv2.THRESH_BINARY)
    if PLOT_THRESH_IMG:
        # Display the original and HSV images side by side
        plt.figure(figsize=(10, 5))
        plt.subplot(1, 2, 1)
        plt.imshow(image)
        plt.title('Original Image')
        plt.subplot(1, 2, 2)
        plt.imshow(thresh_img)
        plt.title('Thresholded Image')
    
    return thresh_img
```
